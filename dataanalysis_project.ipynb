# **This code creates a waterfall chart using Plotly to visualize the contribution of each diagnosis category to the total mood score. It groups data by diagnosis, calculates the sum of mood scores for each, and then displays the results in a visually intuitive format.**

```python
import plotly.graph_objects as go

diagnosis_mood = df_cleaned.groupby('diagnosis')['mood_score_(1-10)'].sum().reset_index()

colors = ['#FF6347', '#90EE90', '#4682B4']

fig = go.Figure(go.Waterfall(
    name="Mood Score by Diagnosis",
    orientation="v",
    measure=["relative"] * (len(diagnosis_mood) - 1) + ["total"],
    x=diagnosis_mood['diagnosis'],
    textposition="outside",
    y=diagnosis_mood['mood_score_(1-10)'],
    connector={"line": {"color": "rgb(63, 63, 63)"}},
    decreasing={"marker": {"color": colors[0]}},
    increasing={"marker": {"color": colors[1]}},
    totals={"marker": {"color": colors[2]}}
))

fig.update_layout(
    title="Mood Score Contribution by Diagnosis Category",
    showlegend=True
)

fig.show()
```

# **This code creates an interactive dashboard using Panel, allowing users to select a therapy type and view a boxplot of mood scores versus sleep quality for the chosen therapy. It dynamically updates the plot based on user input.**

```python
import panel as pn
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

pn.extension('matplotlib')

data = df_cleaned[['sleep_quality_(1-10)', 'mood_score_(1-10)', 'therapy_type']]

therapy_select = pn.widgets.Select(name='therapy_type', options=df_cleaned['therapy_type'].unique().tolist())


@pn.depends(therapy_select.param.value)
def update_plot(therapy_type):
    filtered_data = data[data['therapy_type'] == therapy_type]
    fig, ax = plt.subplots(figsize=(8, 6))
    sns.boxplot(x='mood_score_(1-10)', y='sleep_quality_(1-10)',
                data=filtered_data, ax=ax,
                palette="Set2")
    ax.set_title(f'Sleep Quality vs Mood for {therapy_type}')
    ax.set_xlabel('Mood(scale 1-10)')
    ax.set_ylabel('Sleep Quality')
    plt.close(fig)
    return fig


interactive_plot = pn.Column(
    therapy_select,
    pn.pane.Matplotlib(update_plot, tight=True)
)


interactive_plot.servable()
```

# **This code creates a 3D surface plot using Plotly to represent the frequency of the most common diagnosis based on treatment duration and progress. It calculates and visualizes the maximum diagnosis counts, with hover details providing insights into the associated diagnosis and frequency.**

```python
import plotly.graph_objects as go
import pandas as pd
import numpy as np

x = df_cleaned['treatment_duration_(weeks)']
y = df_cleaned['treatment_progress_(1-10)']


x_unique = x.unique()
y_unique = y.unique()
x_grid, y_grid = np.meshgrid(x_unique, y_unique)


def calculate_z(x_val, y_val):
    filtered_df = df_cleaned[
        (df_cleaned['treatment_duration_(weeks)'] == x_val) &
        (df_cleaned['treatment_progress_(1-10)'] == y_val)]
    diagnosis_counts = filtered_df['diagnosis'].value_counts()


    most_frequent_diagnosis = diagnosis_counts.index[0] if not diagnosis_counts.empty else None
    max_count = diagnosis_counts.max() if not diagnosis_counts.empty else 0

    return max_count, most_frequent_diagnosis

z_grid = np.zeros_like(x_grid, dtype=float)
diagnosis_labels = np.empty_like(x_grid, dtype=object)

for i in range(x_grid.shape[0]):
    for j in range(x_grid.shape[1]):
        z_grid[i, j], diagnosis_labels[i, j] = calculate_z(x_grid[i, j], y_grid[i, j])

fig = go.Figure(data=[go.Surface(
    z=z_grid,
    x=x_grid,
    y=y_grid,
    customdata=diagnosis_labels,
    hovertemplate='Treatment Duration: %{x}<br>Treatment Progress: %{y}<br>Diagnosis: %{customdata}<br>Frequency: %{z}<extra></extra>'
)])

fig.show()
```

```python
import plotly.graph_objects as go
import plotly.figure_factory as ff
import pandas as pd

data = pd.read_csv('mental_health_data.csv')

table_data = [['Diagnosis', 'Average Mood Score', 'Average Sleep Quality']]
for diagnosis in data['Diagnosis'].unique():
    diagnosis_data = data[data['Diagnosis'] == diagnosis]
    avg_mood = diagnosis_data['Mood Score (1-10)'].mean()
    avg_sleep = diagnosis_data['Sleep Quality (1-10)'].mean()
    table_data.append([diagnosis, avg_mood, avg_sleep])

fig = ff.create_table(table_data, height_constant=60)

diagnoses = data['Diagnosis'].unique()
mood_scores = [data[data['Diagnosis'] == diagnosis]['Mood Score (1-10)'].mean() for diagnosis in diagnoses]
sleep_quality = [data[data['Diagnosis'] == diagnosis]['Sleep Quality (1-10)'].mean() for diagnosis in diagnoses]

trace1 = go.Scatter(x=mood_scores, y=sleep_quality,
                    mode='markers',
                    marker=dict(color='#0099ff'),
                    text=diagnoses,
                    name='Diagnosis',
                    xaxis='x2', yaxis='y2')

fig.add_traces([trace1])

fig['layout']['xaxis2'] = {}
fig['layout']['yaxis2'] = {}

fig.layout.xaxis.update({'domain': [0, .5]})
fig.layout.xaxis2.update({'domain': [0.6, 1.]})

fig.layout.yaxis2.update({'anchor': 'x2'})
fig.layout.yaxis2.update({'title': 'Average Sleep Quality'})
fig.layout.xaxis2.update({'title': 'Average Mood Score'})

fig.layout.margin.update({'t': 50, 'b': 100})
fig.layout.update({'title': 'Mental Health Dashboard'})

fig.show()
```

```python

import dash
from dash import dcc, html, Input, Output
import plotly.express as px
import pandas as pd


app = dash.Dash(__name__)

app.layout = html.Div([
    html.H1("Interactive Mental Health Dashboard"),
    html.Button("Update Charts", id="update-button", n_clicks=0),
    dcc.Graph(id='graph-1'),
    dcc.Graph(id='graph-2'),
    dcc.Graph(id='graph-3')
])

@app.callback(
    [Output('graph-1', 'figure'),
     Output('graph-2', 'figure'),
     Output('graph-3', 'figure')],
    [Input('update-button', 'n_clicks')]
)
def update_charts(n_clicks):
    fig1 = px.histogram(df_cleaned, x="treatment_duration_(weeks)", color="diagnosis",
                       color_discrete_sequence=px.colors.qualitative.Pastel,
                       nbins=20,
                       title="Treatment Duration by Diagnosis",
                       labels={"treatment_duration_(weeks)": "Treatment Duration (Weeks)", "diagnosis": "Diagnosis"})
    fig1.update_layout(bargap=0.2)

    fig2 = px.scatter(df_cleaned, x="age", y="mood_score_(1-10)", color="diagnosis",
                      title="Mood Score vs. Age",
                      labels={"age": "Age", "mood_score_(1-10)": "Mood Score", "diagnosis": "Diagnosis"},
                      color_discrete_sequence=px.colors.qualitative.Set2, trendline="ols")

    fig3 = px.box(df_cleaned, x="diagnosis", y="sleep_quality_(1-10)", color="gender",
                  title="Sleep Quality by Diagnosis and Gender",
                  labels={"diagnosis": "Diagnosis", "sleep_quality_(1-10)": "Sleep Quality", "gender": "Gender"},
                  color_discrete_sequence=px.colors.qualitative.Dark24)

    return fig1, fig2, fig3


if __name__ == '__main__':
    app.run_server(debug=True)
```
