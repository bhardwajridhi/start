```python
year_filter = pn.widgets.IntSlider(name='Year', start=int(df['Year'].min()), end=int(df['Year'].max()), value=int(df['Year'].min()))
strata_filter = pn.widgets.Select(name='Strata', options=list(df['Strata'].dropna().unique()), value='Total')
```

```python
@pn.depends(year_filter.param.value, strata_filter.param.value)
def filtered_table(year, strata):
    filtered = df[(df['Year'] == year) & (df['Strata'] == strata)]
    if filtered.empty:
        return "No data available for the selected filters."
    return filtered.hvplot.table()
```

```python
def ci_plot(year):
    year_data = df[df['Year'] == year]
    fig, ax = plt.subplots(figsize=(10, 6))
    if year_data.empty:
        ax.text(0.5, 0.5, f'No data available for the year {year}',
                horizontalalignment='center', verticalalignment='center',
                transform=ax.transAxes, fontsize=14)
    else:
        ax.bar(year_data['Strata Name'], year_data['Percent'], color='skyblue')
        ax.set_title(f'Percent by Strata in {year}')
        ax.set_xticks(range(len(year_data)))
        ax.set_xticklabels(year_data['Strata Name'], rotation=45)
        ax.set_xlabel('Strata Name')
        ax.set_ylabel('Percent')
    plt.tight_layout()
    return fig

```
